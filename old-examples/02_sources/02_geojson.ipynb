# GeoJSON Source

There are several ways to visualize GeoJSON data. For GeoJSON sources, it is not necessary to set the context.

## GeoJSON File

```python
from cartoframes.viz import Map, Layer


Map(Layer('points.geojson'))
```

## GeoJSON full object

```python
geojson = {
    "features": [{
        "type": "Feature",
        "properties": {
            "name": "Coors Field",
            "show_on_map": True
        },
        "geometry": {
            "type": "Point",
            "coordinates": [-104.99404, 39.75621]
        }
    }, {
        "type": "Feature",
        "properties": {
            "name": "Busch Field",
            "show_on_map": False
        },
        "geometry": {
            "type": "Point",
            "coordinates": [-104.98404, 39.74621]
        }
    }]
}

Map(Layer(geojson))
```

## GeoJSON list of features

```python
features = [{
    "type": "Feature",
    "properties": {"party": "Republican"},
    "geometry": {
        "type": "Polygon",
        "coordinates": [[
            [-104.05, 48.99],
            [-97.22,  48.98],
            [-96.58,  45.94],
            [-104.03, 45.94],
            [-104.05, 48.99]
        ]]
    }
}, {
    "type": "Feature",
    "properties": {"party": "Democrat"},
    "geometry": {
        "type": "Polygon",
        "coordinates": [[
            [-109.05, 41.00],
            [-102.06, 40.99],
            [-102.03, 36.99],
            [-109.04, 36.99],
            [-109.05, 41.00]
        ]]
    }
}]

Map(Layer(features))
```

## Single Feature

```python
feature = {
  "type": "Feature",
  "geometry": {
    "type": "Point",
    "coordinates": [125.6, 10.1]
  },
  "properties": {
    "name": "Dinagat Islands"
  }
}

Map(Layer(feature))
```

## GeoJSON geometry

```python
geometry = {
    "type": "LineString",
    "coordinates": [[-100, 40], [-105, 45], [-110, 55]]
}

Map(Layer(geometry))
```
