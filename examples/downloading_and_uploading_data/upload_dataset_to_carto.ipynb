# Upload a Dataset to CARTO

This example illustrates how to upload local data to CARTO

> Note: CARTO Account credentials are needed to reproduce this example. https://carto.com/signup

```python
from cartoframes.auth import set_default_credentials

set_default_credentials('johnsmith', '1a2b3c4d5e6f7g8h9i0j')
```

```python
import geopandas

gdf = geopandas.read_file('../files/sustainable_palm_oil_production_mills.geojson')
gdf.head()
```

```python
from cartoframes.data import Dataset

ds = Dataset(gdf)

ds.upload(
    table_name='sustainable_palm_oil_production_mills',
    with_lnglat=('longitude', 'latitude'),
    if_exists=Dataset.IF_EXISTS_REPLACE
)
```
