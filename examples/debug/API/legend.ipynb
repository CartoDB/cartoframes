```python
from cartoframes.auth import set_default_context
from cartoframes.viz import Map, Layer, Source, Style, Legend

set_default_context(
    base_url='https://cartovl.carto.com/',
    api_key='default_public'
)
```

```python
# Legend color
map = Map(
    Layer(
        Source('populated_places'),
        Style('color: ramp(top($adm0name, 5), bold)'),
        legend=Legend({
            'type': 'color-category',
            'prop': 'color',
            'title': '[TITLE]',
            'description': '[description]',
            'footer': '[footer]'
        })
    )
)

# Legend color + sugar
Map(
    Layer(
        'populated_places',
        'color: ramp(top($adm0name, 5), bold)',
        legend={
            'type': 'color-category',
            'prop': 'color',
            'title': '[TITLE]',
            'description': '[description]',
            'footer': '[footer]'
        }
    )
)
```

```python
# Legend strokeColor
Map(
    Layer(
        'populated_places',
        'color: transparent strokeWidth: 1 strokeColor: ramp(top($adm0name, 5), bold)',
        legend={
            'type': 'color-category',
            'prop': 'strokeColor',
            'title': '[TITLE]',
            'description': '[description]'
        }
    )
)
```

```python
# Legend: only info
Map(
    Layer(
        'populated_places',
        legend={
            'title': '[TITLE]',
            'description': '[description]',
            'footer': '[footer]'
        }
    )
)
```

```python
# Legend: only info
Map(
    Layer(
        'populated_places',
        legend={
            'type': 'color-category',
            'prop': 'color'
        }
    )
)
```

```python
# Legend: complete
Map(
    Layer(
        'SELECT * FROM populated_places WHERE adm0name = \'Spain\'',
        'color: ramp(globalQuantiles($pop_max, 5), reverse(purpor))',
        legend={
            'type': 'color-category',
            'prop': 'color',
            'title': 'Population'
        }
    )
)
```

```python
# Legend size continuous
Map(
    Layer(
        'populated_places',
        """
        width: ramp(buckets($rank_max, [10, 100]), [10, 100])
        strokeWidth: 1
        color: rgba(255, 0, 255, 0.5)
        strokeColor: rgba(0, 0, 0, 0.2)
        """,
        legend={
            'type': 'size-continuous',
            'prop': 'width',
            'title': '[TITLE]',
            'description': '[description]',
            'footer': '[footer]'
        }
    )
)
```

```python
# Legend size bins

Map(
    Layer(
        'populated_places',
        """
        width: ramp($rank_max, [10, 20])
        strokeWidth: 1
        strokeColor: rgba(0, 0, 0, 0.5)
        """,
        legend={
            'type': 'size-bins',
            'prop': 'width',
            'title': '[TITLE]',
            'description': '[description]',
            'footer': '[footer]'
        }
    )
)
```
